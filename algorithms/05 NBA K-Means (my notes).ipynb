Once Upon a time we had some NBA players

```python
import pandas as pd
nba = pd.read_csv("NBA-Census-10.14.2013.csv")
```

And those NBA players all had different positions

```python
nba["POS"].value_counts()
```

And we had decided that they were based on height and weight

```python
for position, marker, color in zip(["C", "F", "G"], 'x>o', 'cmy'):
    players = nba[nba["POS"] == position]
    plt.scatter(players["Ht (In.)"], players["WT"], c=color, alpha=0.5, marker=marker)
```

Although in order to do anything of value we needed to adjust the scores based on standard deviation

```python
nba["adj height"] = (nba["Ht (In.)"] - nba["Ht (In.)"].mean()) / nba["Ht (In.)"].std()
nba["adj weight"] = (nba["WT"] - nba["WT"].mean()) / nba["WT"].std()
```

Because most every formula is based on *distance*, and being 12 pounds away from someone is not nearly as important as 12 inches.

```python
for position, marker, color in zip(["C", "F", "G"], 'x>o', 'cmy'):
    players = nba[nba["POS"] == position]
    plt.scatter(players["adj height"], players["adj weight"], c=color, alpha=0.5, marker=marker)
# There are a dirty-data outliers so we need to set xlim and ylim
plt.xlim(-3,3)
plt.ylim(-3,3)
```

But what if we didn't know the positions of the players?

```python
plt.scatter(nba["adj height"], nba["adj weight"], alpha=0.5)
plt.xlim(-3,3)
plt.ylim(-3,3)
```

Maybe we read in a book that there are three major positions in the NBA, and while we don't even know what the positions are, they're probably based on physical characteristics. Like, say, height and weight.

**k-Means** allows us to guesstimate what these different groups are. Let's start with guessing that there are 3 different groups.

```python
from sklearn.cluster import KMeans

# Turn on verbose to see what it's up to
km = KMeans(n_clusters=3, verbose=True)
```

When k-means attempts to make the model it iterates several times to find the best fit. The default is 10 times.

```python
km.fit(nba[["adj height", "adj weight"]])
```

`km.labels` gives you the labels it has assigned to each element of your dataframe or list.

```python
km.labels_
```

You can either 1) assign that into your dataframe as a new column, or 2) pass the list as a color when plotting. The second one works because 0, 1, 2... etc are all mapped to colors behind the scenes. First, we'll do the latter.

```python
plt.scatter(nba["adj height"], nba["adj weight"], alpha=0.5, c=km.labels_)
plt.xlim(-3,3)
plt.ylim(-3,3)
```

You can also plot the centroids, the point that is being clustered around

```python
km.cluster_centers_
```

We can use `[:,0]` to get all of the X values and `[:,1]` to get all of the Y values, and plot it right on top of the player chart.

```python
plt.scatter(nba["adj height"], nba["adj weight"], alpha=0.5, c=km.labels_)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=200, c=np.unique(km.labels_))
plt.xlim(-3,3)
plt.ylim(-3,3)
```

Let's see what happens if we run it with 5 clusters

```python
km = KMeans(n_clusters=5, verbose=False)
km.fit(nba[["adj height", "adj weight"]])
```

And then plot the results along with the cluster centers

```python
plt.scatter(nba["adj height"], nba["adj weight"], alpha=0.5, c=km.labels_)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=200, c=np.unique(km.labels_))
plt.xlim(-3,3)
plt.ylim(-3,3)
```

...where'd that fifth group go?
