```python
height_distance_from_mean = nba_df["Ht (In.)"] - nba_df["Ht (In.)"].mean()
adjusted_heights = height_distance_from_mean / nba_df["Ht (In.)"].std()
adjusted_heights[:5]
```

```python
nba_df["Adj Height"] = adjusted_heights
```

```python
adjusted_weights = (nba_df["WT"] - nba_df["WT"].mean()) / nba_df["WT"].std()
nba_df["Adj Weight"] = adjusted_weights
```

```python
nba_df[["Name", "WT","Adj Weight","Ht (In.)", "Adj Height"]][:5]
```

```python
training_df = nba_df[:20]

adj_height = (76 - nba_df["Ht (In.)"].mean()) / nba_df["Ht (In.)"].std() 
adj_weight = (220 - nba_df["WT"].mean()) / nba_df["WT"].std()

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(training_df[["Adj Height", "Adj Weight"]], training_df["POS"])
print knn.predict([adj_height, adj_weight])
```

```python
test_df = nba_df[20:]
test_df
```

```python
# Testing data is test_data
# Training is training_data

training_df = nba_df[:20]
test_df = nba_df[20:]

knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(training_df[["Adj Height", "Adj Weight"]], training_df["POS"])

predictions = knn.predict(test_df[["Adj Height", "Adj Weight"]])
predictions[:15]
```

```python
prediction_results = test_df["POS"] == predictions
print len(prediction_results[prediction_results == True])
print len(prediction_results[prediction_results == False])
prediction_results.mean()
```

```python
for training_size in [20, 50, 75, 100, 250, 400, 500, 526, 527]:
    print "Training size: %s" % (training_size)
    training_df = nba_df[:training_size]
    test_df = nba_df[training_size:]

    for k in [1, 2, 3, 4, 5, 10, 15, 20]:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(training_df[["Adj Height", "Adj Weight"]], training_df["POS"])

        predictions = knn.predict(test_df[["Adj Height", "Adj Weight"]])
        prediction_results = test_df["POS"] == predictions
        print "    k=%s: %s" % (k, prediction_results.mean())
```
