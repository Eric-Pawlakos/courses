```python
import pandas as pd
nba_df = pd.read_csv("NBA-Census-10.14.2013.csv")
```

```python
nba_df
```

```python
nba_df.groupby("POS").describe()
```

```python
import matplotlib.pylab as plt
```

```python
plt.scatter(nba_df["Ht (In.)"], nba_df["WT"])
```

```python
centers = nba_df[nba_df["POS"] == 'C']
guards = nba_df[nba_df["POS"] == 'G']
forwards = nba_df[nba_df["POS"] == 'F']
```

```python
plt.scatter(centers["Ht (In.)"], centers["WT"])
```

```python
plt.scatter(guards["Ht (In.)"], guards["WT"])
```

```python
plt.scatter(centers["Ht (In.)"], centers["WT"], c='c', alpha=0.5, marker='x')
plt.scatter(guards["Ht (In.)"], guards["WT"], c='y', alpha=0.5, marker='4')
plt.scatter(forwards["Ht (In.)"], forwards["WT"], c='m', alpha=0.5, marker='>')
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
axs[0,0].set_title("Centers")
axs[0,0].scatter(centers["Ht (In.)"], centers["WT"], c='c')
axs[1,0].set_title("Forwards")
axs[1,0].scatter(forwards["Ht (In.)"], forwards["WT"], c='y')
axs[0,1].set_title("Guards")
axs[0,1].scatter(guards["Ht (In.)"], guards["WT"], c='m')
axs[1,1].set_title("Centers (again)")
axs[1,1].scatter(centers["Ht (In.)"], centers["WT"], c='c')
```

```python
fig, axs = plt.subplots(1, 3, sharex=True, sharey=True)
axs[0].set_title("Centers")
axs[0].scatter(centers["Ht (In.)"], centers["WT"], c='c')
axs[1].set_title("Forwards")
axs[1].scatter(forwards["Ht (In.)"], forwards["WT"], c='y')
axs[2].set_title("Guards")
axs[2].scatter(guards["Ht (In.)"], guards["WT"], c='m')
```

```python
fig, axs = plt.subplots(3, 1, sharex=True, sharey=True)
axs[0].set_title("Centers")
axs[0].scatter(centers["Ht (In.)"], centers["WT"], c='c')
axs[1].set_title("Forwards")
axs[1].scatter(forwards["Ht (In.)"], forwards["WT"], c='y')
axs[2].set_title("Guards")
axs[2].scatter(guards["Ht (In.)"], guards["WT"], c='m')
```
