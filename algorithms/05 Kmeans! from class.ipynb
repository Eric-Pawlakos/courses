```python
nba_df["adj height"] = (nba_df["Ht (In.)"] - 
                        nba_df["Ht (In.)"].mean()) / nba_df["Ht (In.)"].std()
nba_df["adj weight"] = (nba_df["WT"] - 
                        nba_df["WT"].mean()) / nba_df["WT"].std()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, verbose=False)
```

```python
km.fit(nba_df[["adj height", "adj weight"]])
```

```python
km.labels_[:10]
```

```python
nba_df["k-means category"] = km.labels_
nba_df[["POS","k-means category"]][:20]
```

```python
plt.scatter(nba_df["adj height"], nba_df["adj weight"], c=km.labels_)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=100)
```

```python
def salary_to_money(row):
    numbers_only = row['2013 $'].replace(',','').replace('$','')
    if numbers_only == 'n/a':
        return 0
    else:
        return int(numbers_only)
    
nba_df["salary"] = nba_df.apply(salary_to_money, axis=1)
```

```python
nba_df["salary"].hist(bins=500)
```

```python
km = KMeans(n_clusters=5)
```

```python
km.fit(nba_df[["salary"]])
```

```python
nba_df["salary bucket"] = km.labels_
nba_df[["salary","salary bucket"]][:10]
```

```python
nba_df.groupby("salary bucket")["salary"].describe()
```
