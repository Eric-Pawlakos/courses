```python
# First let's import the packages we need
from census import Census
from us import states

# Then we'll initialize a Census object
c = Census("64b41a2cc2325ea98cef7597c1075802e3b8c7c2")

# Then let's use that object to get the number of homes
# that were build before 1939 in the state of New York
c.acs.state(('NAME', 'B25034_010E '), states.NY.fips)

# The code above is the same thing as doing
# c.acs.state(('NAME', 'B25034_010E '), '36')

# If we wanted to get the results for all the states, use
# c.acs.state(('NAME', 'B25034_010E '), Census.ALL)
```

```python
# Let's store all of the Pakistani numbers from every state
pakistani_results = c.acs.state(('NAME', 'B02006_013E'), Census.ALL)
```

```python
# Let's look at it to make sure it's what we think it is
pakistani_results
```

```python
# What's the Pakistani population in the first state?
pakistani_results[0]['B02006_013E']
```

```python
# Let's use a list comprehension to convert every state to an amount
# We have to use int() because the population comes through as a unicode string
[int(result['B02006_013E']) for result in pakistani_results]
```

```python
# Use a lambda to tell max() how to compare the dictionaries
# You need to conver to int() because the population is oddly a string
max(pakistani_results,key=lambda result: int(result['B02006_013E']))
```

```python
# You can do the same thing with min
min(pakistani_results,key=lambda result: int(result['B02006_013E']))
```

```python
# Unfortunately, this only tells us the raw number, not percentage
# Let's also grab the total population (B01001_001E) and try this again
full_results = c.acs.state(('NAME', 'B02006_013E', 'B01001_001E'), Census.ALL)
```

```python
# Calculate the percent of the population that is Pakistani
# You'll need to cast to a float because integer math doesn't do decimals
# Integer: 1 / 4 is 0
# Float: 1.0 / 4.0 is 0.25
for result in full_results:
    result['percent'] =  100 * float(result['B02006_013E']) / float(result['B01001_001E'])
```

```python
# Let's do a sanity check to make sure we stored the percent correctly
full_results
```

```python
# Now we can sort by the percent key
max(full_results,key=lambda result: result['percent'])
```

```python
# Once again, this doesn't work! max() doesn't know how 
# to compare the results to one another
max(full_results)
```

```python
# If we want more than the max and the min, we can use sorted()
# to get them all in order
sorted(full_results,key=lambda result: result['percent'])
```

```python
# Instead of doing c.acs.state, we can do c.acs.state_county to get each county
# We just have to pass it the fields, then the FIPS code of the state, then the counties
# we want. So, '36' is the FIPS code for New York, and Census.ALL is every single county
county_results = c.acs.state_county(('NAME', 'B02006_013E', 'B01001_001E'), '36', Census.ALL)
```

```python
# Let's bask in the results
county_results
```

```python
# Now we can do the same thing as before to calculate the  
# percent of the total population
for result in county_results:
    result['percent'] =  100 * float(result['B02006_013E']) / float(result['B01001_001E'])
    
sorted(county_results,key=lambda result: result['percent'])
```
